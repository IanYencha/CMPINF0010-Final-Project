Looking at Food Gardens in Pittsburgh

```python
# All Required Imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Original Database. Way to many collumns that I don't need
full_trees = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da")
# Shortened database I will be using
trees = full_trees[['common_name', 'height','condition','neighborhood','air_quality_benfits_total_dollar_value']].copy()
trees.head(10)
```

```python
print("There are " + str(len(trees)) + " trees managed by the City of Pittsburgh")
trees.describe()
```

```python
# BE ADVISED: This code block takes some time to run (15-30 sec)
count_dict = {}
x = 0
while x < len(trees):
    if trees.iloc[x]['neighborhood'] in count_dict.keys():
        count_dict[trees.iloc[x]['neighborhood']] =  count_dict.get(trees.iloc[x]['neighborhood']) + 1
    else:
        count_dict.update({trees.iloc[x]['neighborhood'] : 1})
    x += 1
#print(count_dict)

# Goal is to get a count of how many trees per neighborhood. 
# next step is converting to dataframe
```

```python
#Takes dictionary from code cell above and translates into DataFrame
trees_per_neighborhood = pd.DataFrame([count_dict])
trees_per_neighborhood = trees_per_neighborhood.T
trees_per_neighborhood = trees_per_neighborhood.rename(columns = {0 : "number_of_trees"})

trees_per_neighborhood.sort_values(by = 'number_of_trees', ascending = False).head(10)
```
